```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import datetime
import numpy as np
```

```python
df = pd.read_csv("raw.csv",parse_dates=["Timestamp"])
```

```python
df.head(20)
```

# Adding Year , Month , Day , Weekday and hour


```python

df["year"] = df["Timestamp"].dt.year
df["month"] = df["Timestamp"].dt.month
df["day"] = df["Timestamp"].dt.day
df["weekday"] = df["Timestamp"].dt.day_name()  
df["hour"] = df["Timestamp"].dt.hour
```

# Adding features from weather data to the main dataset

```python
df1 = weather.sort_values("valid_time")  # Ensure sorted by valid_time
df2 = df.sort_values("Timestamp")   # Ensure sorted by timestamp

df = pd.merge_asof(
    df2, 
    df1, 
    left_on="Timestamp", 
    right_on="valid_time",  
    direction="backward"
)
```
